**Introduction to matplotlib**  
This notebook introduces the basic knowledge about matplotlib.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('classic') # choose the kind of plot style

# a magic function to lead to static images in the notebook
%matplotlib inline 
```

```python
# find the different style you can use
print(plt.style.available)
```

```python
import numpy as np

x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure()
plt.plot(x,y1,'-') # '-' means a solid line 
plt.plot(x,y2,'--') # '--' means a dash line
```

```python
# save figures to file
fig.savefig('sin(x)_and_cos(x).jpg')
```

```python
# find the type of files we can save as
fig.canvas.get_supported_filetypes()
```

```python
# use IPython module to display the picture
from IPython.display import Image
Image('sin(x)_and_cos(x).jpg')
```

```python
# plot two panels in one canvas
plt.figure() # create a plot canvas

# create two panels in 2 rows and 1 column
plt.subplot(2,1,1)
plt.plot(x,y1)

plt.subplot(2,1,2)
plt.plot(x,y2)
```

```python
# get the current figure and axes
plt.gcf()

plt.gca()
```

```python
# another method to plot two panels in one canvas
fig, ax = plt.subplots(2)

ax[0].plot(x,y1)
ax[1].plot(x,y2)
```
