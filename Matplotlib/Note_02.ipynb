```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

**As far as my concerned, the figure represents canvas and the axes represent the drawing.**

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
y = np.sin(x)
ax.plot(x,y,'-g',linewidth=2)
```

```python
plt.plot(x,y,'-b',linewidth=2)
```

```python
# differnet styles and colors of plots
plt.plot(x,x+0,'-g') # solid green
plt.plot(x,x+2,'--c') # dash cyan
plt.plot(x,x+4,'-.k') # dashdot black
plt.plot(x,x+6,':r') # dotted red
```

```python
# set the axes limits
plt.plot(x,np.sin(x))
plt.xlim(-1,10)
plt.ylim(-1,1)
```

```python
# you can also use plt.axis()
plt.plot(x,np.sin(x))
plt.axis([-1,10,-1,1])
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

```python
# set labels and legend
plt.plot(x,np.sin(x),'--g',linewidth=3,label='sin(x)')
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
```

```python
# you can also set axes
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-1.5,1.5),
      xlabel='x',ylabel='sin(x)',
      title='A Sine Curve')
```

- ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``
